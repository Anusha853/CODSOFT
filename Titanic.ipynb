```python
#Loading the dataset
import pandas as pd
df=pd.read_csv("titanic.csv")
df
```

```python
## PREPROCESSING
```

```python
df.isnull().any() #null values in Age,Fare and Cabin attributes
```

```python
df.isnull().sum()
```

```python
df.Age.fillna(df.Age.median(),inplace=True) #replacing null values with median since it has numerical data
df.Fare.fillna(df.Fare.median(),inplace=True) #replacing null values with median since it has numerical data
df=df.drop("Cabin",axis=1) #Because Cabin has many null values
```

```python
df.isnull().any() #No null values in the dataset
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder() #preferred when there are less no.of categories for attribute
```

```python
df.Sex=le.fit_transform(df.Sex)
df.SibSp=le.fit_transform(df.SibSp)
df.Parch=le.fit_transform(df.Parch)
df.Embarked=le.fit_transform(df.Embarked)
```

```python
df
```

```python
##Identifying dependent and independent variables in the dataset
X=df.iloc[::]
X=X.drop('Survived',axis=1)
X=X.drop('Name',axis=1) #Name attribute doesn't effect on the result
X=X.drop('Ticket',axis=1) #Ticket attribute doesn't effect on the result

y=df["Survived"] # The independent Variable
```

```python
X.head()
```

```python
y
```

```python
#Scaling the data
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
```

```python
x_scaled=pd.DataFrame(scale.fit_transform(X),columns=X.columns)
x_scaled.head()
```

```python
##Splitting the data into train and test data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Building the model and training
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,precision_score,recall_score
```

```python
#Decision Tree Classifier
clf = DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#KNN Classifier
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("\nClassification Report:")
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```

```python
#Naive Baye's Classifier
clf = GaussianNB()
clf = clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
precision = precision_score(y_test, y_pred, average='micro',pos_label=2)
print('Precision: ',precision)
recall = recall_score(y_test, y_pred, average="micro",pos_label=2)
print('Recall: ',recall)
```
